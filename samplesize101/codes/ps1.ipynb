## Q1: 検出力曲線のplot
とあるコインの表が出る確率をPで表し、そのサポートを
$$
P \in (0, 1)
$$

n回投げてそのうち表が出た回数をxで表す。

このとき、
$$
Pr(x = r) = _nC_rP^r(1-P)^{n-r}
$$

$H_0: P = 0.5$, $H_1: P \neq 0.5$とする

### 手順
1. $\alpha_0$を設定
2. $p = 0.5$と$\alpha = \alpha_0$のときの帰無仮説棄却水準となる$r_{lower}$と$r_{upper}$を求める
3. True probabilityのもと、$Pr(r_{lower}< r < r_{upper})$を求める

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
## probability mass function
def pmf(n, p, r):
    """
    function : 二項分布のprobability mass function
    
    Input
        n : sample size
        p : probability
        r : headの回数
    """
    
    _first_term = math.factorial(n)/(math.factorial(r)*math.factorial(n-r))
    _second_term = p**(r)*(1-p)**(n-r)
    
    return _first_term * _second_term

```

```python
## 帰無仮説棄却水準のrを導出
def significant_level(n, p, threshold, lower = True):
    """
    function : 帰無仮説棄却水準のrを導出
    
    Input
        n : sample size
        p : probability
        threshold : 検定水準
    """
    eps = 10e-7 ## 計算機イプシロン
    
    if lower:
        _cdf = 0
        _r = -1
        while _cdf <= threshold and _r <= n:
            _r += 1
            _cdf += pmf(n, p, _r)
        
        if _cdf > threshold + eps:
            _r -= 1
            
    else:
        _r = n+1
        _cdf = 0
        _threshold = 1 - threshold
        while _cdf <= _threshold and _r >= 0:
            _r -= 1
            _cdf += pmf(n, p, _r)
        
        if _cdf > _threshold + eps:
            _r += 1
        
    return _r
```

```python
## cumulative distribution function
def cdf(n, p, threshold):
    """
    function : cumulative districution function
    
    Input
        n : sample size
        p : probability
        threshold : 検定水準
    """
    _r = 0
    _cdf = 0
    while _r < threshold:
        _cdf += pmf(n, p, _r)
        _r += 1
    
    return _cdf
```

```python
### main関数
def power_analysis(sample_size, p_1 , p_0, alpha):
    r_lower = significant_level(n = sample_size, p = p_0, threshold = alpha, lower = True)
    r_upper = significant_level(n = sample_size, p = p_0, threshold = alpha, lower = False)
    
    beta = cdf(n = sample_size, p = p_1, threshold = r_upper) - cdf(n = sample_size, p = p_1, threshold = r_lower)
    
    return beta
    
    
```

```python
### 検出力曲線のplot
p_array = np.linspace(0, 1, 100)[1:-1] ### Truth
p_0 = 0.5 ## H_0

alpha_1 = 0.05 ## significance level
alpha_2 = 0.1
alpha_3 = 0.2

sample_size = 20 ## sample size
```

```python
plt.figure(figsize = (10, 10))

for alpha_0 in (alpha_1, alpha_2, alpha_3):
    y = []
    for p in p_array:
        beta = 1 - power_analysis(sample_size, p_1 = p , p_0 =p_0, alpha = alpha_0)
        y.append(beta)
    
    plt.plot(p_array, y, label = 'alpha = {}'.format(alpha_0))
    
plt.xlabel('true probability')
plt.ylabel('power analysis')
plt.legend();
```
