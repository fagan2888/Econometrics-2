### Q3: norm cdfをスクラッチから書け

```python
import math
```

```python
def erf(x):
    ''' John D. Cook's implementation.http://www.johndcook.com
    >> Formula 7.1.26 given in Abramowitz and Stegun.
    >> Formula appears as 1 – (a1t1 + a2t2 + a3t3 + a4t4 + a5t5)exp(-x2)
    >> A little wisdom in Horner's Method of coding polynomials:
        1) We could evaluate a polynomial of the form a + bx + cx^2 + dx^3 by coding as a + b*x + c*x*x + d*x*x*x.
        2) But we can save computational power by coding it as ((d*x + c)*x + b)*x + a.
        3) The formula below was coded this way bringing down the complexity of this algorithm from O(n2) to O(n).'''
 
    # constants
    a1 =  0.254829592
    a2 = -0.284496736
    a3 =  1.421413741
    a4 = -1.453152027
    a5 =  1.061405429
    p  =  0.3275911
 
    # Save the sign of x
    sign = 1
    if x < 0:
        sign = -1
    x = abs(x)
 
    # Formula 7.1.26 given in Abramowitz and Stegun.
    t = 1.0/(1.0 + p*x)
    y = 1.0 - (((((a5*t + a4)*t) + a3)*t + a2)*t + a1)*t*math.exp(-x*x)
 
    return sign*y
```

```python
def phi(x):
    '''Cumulative gives a probability that a statistic
    is less than Z. This equates to the area of the
    distribution below Z.
    e.g:  Pr(Z = 0.69) = 0.7549. This value is usually
    given in Z tables.'''
 
    return 0.5*( 1.0 + erf(x/math.sqrt(2)) )
```

```python
### universal 函数を用いて表現
_phi = np.vectorize(phi)
plt.plot(bins, _phi(bins), label = 'self-made')
plt.plot(bins , norm.cdf(bins), label = 'scipy')
plt.title('normal distribution')
plt.legend()
```

### Q5: 正規分布の再生性の確認
$$
X \sim N(\mu_x, \sigma^2_x)
$$
$$
Y \sim N(\mu_y, \sigma^2_y)
$$
かつ、X, Yが独立のとき
$$
X + Y \sim N(\mu_x + \mu_y, \sigma^2_x + \sigma^2_y)
$$

### Proof
$$
\begin{aligned}
M_{x+y}(t) & = E[\exp(t(x+y))]\\
& = E[\exp(tx)\exp(ty)]\\
& = E[\exp(tx)]E[\exp(ty)]\\
& = \exp\left(\mu_x + \frac{\sigma_x^2t^2}{2}\right)\exp\left(\mu_y + \frac{\sigma_y^2t^2}{2}\right)\\
& = \exp\left(\mu_x + \mu_y + \frac{(\sigma_x^2+\sigma^2_y)t^2}{2}\right)
\end{aligned}
$$

```python
mu_x = 3
mu_y = 7
sigma_x = 2
sigma_y = 4

x = np.random.normal(mu_x, sigma_x, 10000)
y = np.random.normal(mu_y, sigma_y, 10000)

z = x + y
mu_z = mu_x + mu_y
sigma_z = np.sqrt(sigma_x **2 + sigma_y **2)

count, bins, ignored = plt.hist(z, 30, density=True)
plt.plot(bins, 1/(sigma_z * np.sqrt(2 * np.pi)) * 
         np.exp( - (bins - mu_z)**2 / (2 * sigma_z**2)),
         linewidth=2, color='r');

```

```python
print("mean = {}, var ={}".format(round(np.mean(z),3),
                                  round(np.var(z), 3)))
```

#### 6-1

##### (1) $(a, b)$上の一様分布
$$
f(x) = \begin{cases}
c & x \in (a, b)\\
0 & \text{otherwise}
\end{cases}
$$
##### 解
$$
c = \frac{1}{b - a}
$$
さらに平均と分散は
$$
M(t) = \frac{1}{b - a}\frac{1}{t}[\exp(tb) - \exp(ta)]
$$
Then, 
$$
\mu = \frac{1}{2}(b + a)
$$
$$
var = \frac{1}{12}(b - a)^2
$$

```python
self_mean = (a + b)/2
self_var = round((b - a)**2/12, 2)
```

```python
print("sample_mean = {}, sample_var = {}".format(round(np.mean(x_unif), 2), round(np.var(x_unif), 2)))
print("pop_mean = {}, pop_var = {}".format(self_mean
                                            , self_var))
```

f(x) = \begin{cases}
c(x^2 + 2) & x \in (0, 3)\\
0 & \text{otherwise}

#### 6-2 : 共分散の性質を示せ
(1) covariance

$$
\begin{aligned}
cov(x, y)  & = E[(x - \mu_x)(y - \mu_y)]\\
& = E[xy] - E[x]\mu_y - E[y]\mu_x + \mu_x\mu_y\\
& = E[xy] - \mu_x\mu_y
\end{aligned}
$$

(2) variance
a, b を定数とする
$$
\begin{aligned}
V(ax + by) & = E[(ax + by - a\mu_x - b\mu_y)^2]\\
& = aE[(x-\mu_x)^2] + bE[(y -\mu_y)^2] + 2ab E[(x- \mu_x)(y - \mu_y)]\\
& = aVar(x) + bVar(y) + 2abCov(x,y)
\end{aligned}
$$


#### 6-3: 相関係数について各設問に答えよ
##### (1)
$$
\begin{aligned}
\mu & = E[x]\\
\mu_y & = E[y]\\
g(t) & = E[\{(x - \mu_x) - t(y - \mu_y)\}^2]
\end{aligned}
$$

##### 解
$$
\begin{aligned}
g(t) & = E[(x - \mu_x)^2] + t^2 E[(y - \mu_y)^2] - 2tE[(x - \mu_x)(y - \mu_y)]\\
& = V(x) + t^2 V(y) - 2tCov(x, y)
\end{aligned}
$$

##### (2) 
$g(t)$がtの二次関数であるとして、$g(t) \geq 0$となる条件を求めよ

##### 解
$$
Cov(x, y)^2 - V(x)V(y)\leq 0 \Rightarrow cov(x, y)^2\leq V(x)V(y)
$$


#### 6-4 : 同時確率密度関数

##### (1)

##### 解
$$
\begin{aligned}

\end{aligned}
$$
Then, 
$$
c = 4
$$


##### (2)
x, y それぞれの周辺密度関数を求めよ

##### 解
$$
\begin{aligned}
f_x(x) & = 2x\\
f_y(y) & = 2y
\end{aligned}
$$


##### (3)
x,yが互いに独立であることを示せ

##### 解
$$
f(x, y) = f_x(x)f_y(y)
$$

##### (4)
$E(x)$及び$E[2x^2y]$を求めよ

##### 解
$$
E[x] = \int^1_0 xf(x)dx = \int^1_0 2x^2 dx = \left[\frac{2}{3}x^3\right]^1_0 = \frac{2}{3}
$$
また
$$
\begin{aligned}
E[2x^2y] & = 2*E[x^2]E[y]\\
& = 2\cdot\frac{2}{3}\cdot \frac{1}{2}\\
& = \frac{2}{3}
\end{aligned}
$$

#### 6-6 : inverse CDF

```python
def formula(t):
    #constants
    c0 = 2.515517
    c1 = 0.802853
    c2 = 0.010328
    d0 = 1.432788
    d1 = 0.189269
    d2 = 0.001308  
 
    # Formula
    p = t - ((c2*t + c1)*t + c0) / (((d2*t + d1)*t + d0)*t + 1.0)
    return  p
 
def q(p):
    if (p < 0.5):
        #F^-1(p) = - G^-1(p)
        return -formula( math.sqrt(-2.0*math.log(p)) )
 
    else:
        #F^-1(p) = G^-1(1-p)
        return formula( math.sqrt(-2.0*math.log(1-p)) )
 
    return q
```

#### 6-9: 2次元正規分布の同時密度関数
$$
f(x, y) = \frac{1}{2\pi \sqrt{1 - \rho^2_{xy}}\sigma_x\sigma_y}\exp\left(-\frac{1}{2}\frac{1}{1 - \rho^2_{xy}}\left[\frac{(x - \mu_x)^2}{\sigma_x^2} - 2\rho_{xy}\frac{(x - \mu_x)(y - \mu_y)}{\sigma_x\sigma_y} + \frac{(y - \mu_y)^2}{\sigma_y^2}\right]\right)
$$

ここから周辺密度関数$f(x)$を求めよ

##### 解
$$
f(x) = \int^\infty_{-\infty} f(x, y)dy
$$
ここで
$$
\begin{aligned}
(1 - \rho^2_{xy})D^2 & = \left(\frac{(y - \mu_y)}{\sigma_y} - \rho_{xy}\frac{(x - \mu_x)}{\sigma_x} \right)^2 - \rho_{xy}^2\frac{(x - \mu_x)^2}{\sigma_x^2} + \frac{(x - \mu_x)^2}{\sigma_x^2} \\
& = \left(\frac{(y - \mu_y)}{\sigma_y} - \rho_{xy}\frac{(x - \mu_x)}{\sigma_x} \right)^2 + (1 - \rho_{xy}^2)\frac{(x - \mu_x)^2}{\sigma_x^2}
\end{aligned}
$$


よって
$$
\begin{aligned}
\int f(x, y)dy &= \frac{1}{\sqrt{2\pi}\sigma_x}\exp\left(- \frac{(x - \mu_x)^2}{2\sigma_x^2}\right)\int \frac{1}{\sqrt{2\pi}\sigma_y\sqrt{1 - \rho^2_{xy}}}\exp\left(- \frac{1}{2(1- \rho_{xy}^2)\sigma_y^2}\left(y - \mu_y - \rho_{xy}\frac{\sigma_y}{\sigma_x}(x - \mu_x)\right)^2\right)dy\\
& = f(x)
\end{aligned}
$$

```python
1 - round(norm.cdf(10/np.sqrt(38)), 4)
```

```python
def self_ecdf(x, value):
    """
    input : 
        x: array
        value : cdfのvalue
    return : cdf
    """
    n = len(x)
    tmp = len(x[x < value])
    
    return tmp/n
    
    
```

```python
round(1 - self_ecdf(y, 110) , 4)
```
